```python
#convert below thousand 0-999, three numbers, hundreds, tens and ones
def convert_below_thousand(n):
    if n < 100:
        return convert_below_hundred(n)
    hundreds_part = ONES[n // 100]
    tens_part = convert_below_hundred(n % 100)
    return f"{hundreds_part} hundred and {tens_part}" if tens_part else f"{hundreds_part} hundred"
```

Now the big one — thousands, millions, billions. English groups numbers in threes: "66,723,107,008" is spoken as four groups. So I need to break any number into groups of 3 digits from right to left, convert each group, and attach the right scale word.

The groups are right, now I need to join them properly. Looking at the expected outputs:
- "nine thousand**,** one hundred and twenty-one" — comma between groups
- "ten thousand **and** twenty-two" — "and" before the last group when it's below 100

So the rule is: if the last group (ones position) is less than 100, use "and" instead of a comma to connect it.

```python
def convert(n):
    if n == 0:
        return "zero"

    groups = []
    while n > 0:
        groups.append(n % 1000)
        n //= 1000

    SCALES = ["", "thousand", "million", "billion", "trillion"]

    parts = []
    for group, scale in zip(groups, SCALES):
        if group != 0:
            word = convert_below_thousand(group)
            parts.append(f"{word} {scale}".strip())

    parts.reverse()

    if len(parts) > 1 and 0 < groups[0] < 100:
        return ', '.join(parts[:-1]) + ' and ' + parts[-1]

    return ', '.join(parts)
```

## Part 2: Extracting numbers from text

Now the other half — given a line like `"The pump is 536 deep underground."`, I need to pull out `536`. But `"#65678"` should be invalid and `"23 456,9"` should also be invalid.

Let me explore what regex gives me for each test case.

```python
import re

test_lines = [
    "The pump is 536 deep underground.",
    "We processed 9121 records.",
    "Variables reported as having a missing type #65678.",
    "Interactive and printable 10022 ZIP code.",
    "The database has 66723107008 records.",
    "I received 23 456,9 KGs.",
]

# find all digit sequences in each line
{line: re.findall(r"\d+", line) for line in test_lines}
```

So `"23 456,9"` gives 3 matches — easy to reject (multiple digit groups = ambiguous).

But `"#65678"` gives exactly 1 match. The problem is it's glued to `#`, making it a code/identifier, not a number. I need to check that the number is **standalone** — surrounded by whitespace or sentence punctuation, not attached to special characters.

```python
# need a "standalone" check — number must be surrounded by whitespace/punctuation
STANDALONE_NUMBER = re.compile(r"(?:^|(?<=\s))(\d+)(?=\s|[.,;:!?]|$)")

# check both: how many \d+ matches, and whether the standalone regex finds it
{line: {
    "digits": re.findall(r"\d+", line),
    "standalone": (m := STANDALONE_NUMBER.search(line)) and m.group(1)
} for line in test_lines}
```

The two checks together cover all cases:
1. Exactly one `\d+` match in the line (rejects `"23 456,9"`)
2. That match must be standalone (rejects `"#65678"`)

```python
def extract_number(line):
    all_digits = re.findall(r"\d+", line)
    if len(all_digits) != 1:
        return None

    match = STANDALONE_NUMBER.search(line)
    if not match:
        return None

    return int(match.group(1))

{line: extract_number(line) for line in test_lines}
```

## Part 3: Putting it all together

```python
def process_line(line):
    number = extract_number(line)
    if number is None:
        return "number invalid"
    return convert(number)

expected_outputs = [
    "five hundred and thirty-six",
    "nine thousand, one hundred and twenty-one",
    "number invalid",
    "ten thousand and twenty-two",
    "sixty-six billion, seven hundred and twenty-three million, one hundred and seven thousand and eight",
    "number invalid",
]

results = {line: process_line(line) for line in test_lines}

for (line, result), expected in zip(results.items(), expected_outputs):
    assert result == expected, f"FAIL: got '{result}', expected '{expected}'"

results
```
